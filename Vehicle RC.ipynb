```python
def get_data_image(file, num, imgtype = "image"):
    img = "image"
    if "image" in imgtype:
        img = Image.open(file)
    elif "cv2" in imgtype:
        img = cv2.imread(file)
    text = tess.image_to_string(img)
    data = {"photonumber":str(num)}
    for i in text.split("\n"):
        try:
            key = i.split(":")[0].replace(" ","").lower()
            val = i.split(":")[1].replace(" ","")
            val = val.replace("COLOUR","")
            val = val.replace("CLASS","")
            if "regn.no" in key:
                val = val[:9]
            if "no" in key and len(val)>=17:
                key = "ch.no"
                val = val[:18]
            if "eno" in key:
                key = "eno"
            if "ch.no" in key:
                key = "ch.no"
            if "mfg" in key:
                key = "mfg.dt"
            if "name" in key:
                key = "name"
            if "dt" in key:
                val = val.replace("/","")
                val = val[:8]
                val = val[:2] + "/" + val[2:4] + "/" + val[4:]
            if "reg.dt" in key  or "regndt" in key:
                key = "reg.dt"
            if key != "":
                data[key] = val
        except:
            pass
    return data
```

```python
cv2_images = []
images = []
for index in range(len(files)):
    cv2_images.append(get_data_image(dataDir+files[index],index,"cv2"))
    images.append(get_data_image(dataDir+files[index],index))
```
